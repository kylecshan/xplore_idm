```python
import numpy as np
import h5py
import os
```

```python
from osgeo import gdal
def load_file(folder, file):
    ds = gdal.Open(os.path.join(folder, file))
    return ds.ReadAsArray()
```

```python
IMGFOLDER = 'E:/xplore_data/data/'
path = 'E:/xplore_data/data/images.h5'
K = 833
```

```python
os.remove(path)
imglist = os.listdir(IMGFOLDER)
imglist.sort()
n = len(imglist)
```

```python
with h5py.File(path, "a") as f:
    dset = f.create_dataset('satellite', (n, 10, K, K), dtype='u1')
    cluster_ids = f.create_dataset('cluster_id', (n,), dtype='u2')
    
    for i, file in enumerate(imglist):
        if file == 'images.h5':
            continue
        print(file, end='; ')
        filepath = os.path.join(IMGFOLDER, file)
        array = load_file(IMGFOLDER, file)
        hstart = (array.shape[1]-K)//2
        wstart = (array.shape[2]-K)//2
        output = array[:, hstart:(hstart+K), wstart:(wstart+K)]
        c_id = int(file[-7:-4])
        print(c_id)
        dset[i] = output
        cluster_ids[i] = c_id
```
